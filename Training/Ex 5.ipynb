### Logistic Regression on Bank Marketing data

For the Bank Marketing data set available in <strong>Exercises/data </strong>folder, build a model using logistic regression model to predict the attribute “y” (i.e. the client will subscribe a term deposit or not) based on duration, age, and campaign as predictors.
<br/>Use 70% of the data as training data set and 30% of data as testing data set. 
<br/>Determine the classification accuracy rate of the model on both train and test set.
<br/><b>bank_marketing.csv</b> file contains the data and it's description is available in <b>bank_marketing.text</b> file.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
bank = pd.read_csv("data/bank_marketing.csv",delimiter=";")
```

```python
bank.info()
```

```python
X = bank[["duration","age","campaign"]]
y = bank["y"]
```

```python
bank["duration"]
```

```python
bank.groupby("campaign").count()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
model = LogisticRegression()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
ytrain.shape
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
ypredtest = model.predict(xtest)
ypredtrain = model.predict(xtrain)
```

```python
confusiontest=confusion_matrix(ytest,ypredtest)
confusiontrain=confusion_matrix(ytrain,ypredtrain)
```

```python
print(confusiontest)
```

```python
pd.DataFrame(confusiontest,columns=model.classes_,index=model.classes_)
```

```python
pd.DataFrame(confusiontrain,columns=model.classes_,index=model.classes_)
```

```python
print(classification_report(ytrain,ypredtrain))
```

```python
print(classification_report(ytest,ypredtest))
```

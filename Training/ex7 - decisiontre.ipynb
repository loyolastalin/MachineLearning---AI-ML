### Decision Tree on Bank Marketing data

For the  Bank Marketing dataset available in <strong>Exercises/data </strong>folder, induct a decision tree to predict whether a client will subscribe to a term deposit or not. 
<li>Split the data into train and test with a ratio of 80:20</li>
<li>Do not use 'duration' as a predictor</li>
<b>bank_marketing.csv</b> file contains the data and it's description is available in <b>bank_marketing.text</b> file.

```python
# Write your code here

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bank = pd.read_csv("data/bank_marketing.csv", delimiter=";")
bank.head()
```

```python
features = bank.columns.drop(["duration","y"])
y = bank["y"]
print(features)
```

```python
X= bank[features]
X.head()
```

```python
xencoded = pd.get_dummies(X)
xencoded.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(xencoded,y,test_size = 0.2, random_state = 99)
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtrain,ytrain)
```

```python
ypredtest = model.predict(xtest)
ypredtrain = model.predict(xtrain)
```

```python
conftrain = confusion_matrix(ytrain, ypredtrain)
conftest = confusion_matrix(ytest, ypredtest)
```

```python
pd.DataFrame(conftrain, columns=model.classes_, index=model.classes_)
```

```python
pd.DataFrame(conftest, columns=model.classes_, index=model.classes_)
```

```python
print(classification_report(ytrain, ypredtrain))
```

```python
print(classification_report(ytest, ypredtest))
```
